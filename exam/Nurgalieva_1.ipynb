```python
import scipy.stats as sps
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
N = 1000
K = 10000
theta = 0
X = np.array(sps.norm.rvs(size=N*K, loc=0, scale=1))
X = X.reshape(N, K)
```

```python
# оценка матожидания - это X-среднее
theta_likelyhood = np.array([X[:n].mean(axis=0) for n in range(1, N, 1)])
```

```python
# вторая оценка мат ожидания
theta_hat = np.array([(X[:n].max(axis=0)+X[:n].min(axis=0))*1./2 for n in range(1, N, 1)])
```

```python
# оценим дисперсию
sigma_likelyhood = np.array([theta_likelyhood[:n].var() for n in range(1, N, 1)])
print(sigma_likelyhood[0])
```

```python
sigma_hat = np.array([theta_hat[:n].var() for n in range(1, N, 1)])
```

```python
# построим графики
plt.figure(figsize=(20,10))    
plt.title(r'$\theta \ estimations$', fontsize = 50)
plt.plot(np.arange(1, N), sigma_likelyhood , color = 'blue', label=r'$maxlikelyhood$')
plt.plot(np.arange(1, N), sigma_hat , color = 'red', label=r'$max + min$')
plt.legend(loc='center left', bbox_to_anchor=(0.7, 0.7), fontsize = 30)
plt.ylim(0, 0.4)
plt.show()
```
