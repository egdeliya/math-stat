```python
import scipy.stats as sts
import scipy.misc as sm
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import matplotlib.pyplot as plt
%matplotlib inline
```

Генерируем выборку из стандартного нормального распределения.

```python
N = 1000
samples = sts.norm.rvs(size=N, loc=0, scale=1)
```

Для получения доверительных интервалов будем использовать гауссовскую линейную модель.
Так как у нас $X_1 .. X_n \sim \mathcal{N}(a, \sigma ^2)$, то в нашей модели $Z = (1, .., 1)^T$, так как $l = (a, ..,a)^T$.   

Тогда $\hat{\theta} = (Z^TZ)^{-1} Z^TX$ - клевая оценка $\theta$

#### Строим доверительный интервал для $\sigma^2$ при неизвестном $a$:

Построим доверительный интервал для $\sigma^2$, мы знаем, что
$$\frac{1}{\sigma ^ 2} |X - Z\hat{\theta}|^2 \sim \mathcal{X}^2(n - 1)$$

Значит, $$P\left(0 < \sigma^2 < \frac{|X - Z\hat{\theta}|^2}{z_{1 - \alpha}}\right) = \alpha$$
, где $\alpha$ - уровень доверия


1) Находим квадтилью уровня 0.05 у $\mathcal{X}^2(n - 1)$

2) Ищем наш интервал

```python
alpha = 0.95
```

Подсчет интервала:

```python
def leastSquaresSigma(samples, n):
    x = np.matrix(samples).T
    
    chi = sts.chi2(df=(n - 1.))
    kvan = chi.ppf(1 - alpha)
    
    z = [1 for i in range(n)]
    z = np.matrix(z).T
    
    thetaHat = np.linalg.inv(z.T * z) * z.T * x
    return (np.linalg.norm(x - z * thetaHat) ** 2) / kvan
```

Подсчет интервалов для всех $n$: 

```python
y0 = [0 for i in range(N)]
y1 = []
for n in range(1, N + 1):
    y1.append((leastSquaresSigma(samples[:n], n)))
```

Строим график:

```python
nrange = np.arange(1, N + 1)
plt.fill_between(nrange, y0, y1)
plt.plot(nrange, (lambda y: 1 * y / y)(nrange), color='red')
plt.xlim(0, N)
plt.ylim(0, 5)
plt.show()
```

#### Теперь построим доверительный интервал $a$ при неизвестном $\sigma^2$

Из $$\frac{\hat{a} - a}{\sigma \sqrt{a_{ii}}} \sim \mathcal{N}(0, 1)$$
, где $a_{ii}$ - элемент матрицы $(Z^TZ)^{-1}$
и из $$\frac{1}{\sigma ^ 2} |X - Z\hat{a}|^2 \sim \mathcal{X}^2(n - 1)$$
получаем, что:

$$\frac{\hat{a} - a}{\sqrt{\frac{a_{ii}}{n - 1}}|X - Z\hat{a}|^2} \sim T_{n - 1}$$
, где Т - распределение Стьюдента

Итого:
$$P\left(\hat{a} - z_{\frac{1 + \alpha}{2}} \sqrt{\frac{a_{ii}}{n - 1}} |X - Z\hat{a}|^2 < a < \hat{a} - z_{\frac{1 - \alpha}{2}} \sqrt{\frac{a_{ii}}{n - 1}} |X - Z\hat{a}|^2 \right) = \alpha$$

Подсчет интервала:

```python
def leastSquaresA(samples, n):
    x = np.matrix(samples).T
    #квантили
    stud = sts.t(df=n-1)
    z1 = stud.ppf(1. / 2. - alpha / 2.)
    z2 = stud.ppf(1. / 2. + alpha / 2.)
    #Z
    z = [1 for i in range(n)]
    z = np.matrix(z).T
    #a_ii
    a_ii = np.array(np.linalg.inv(z.T * z))[0][0]
    #aHat
    aHat = np.linalg.inv(z.T * z) * z.T * x
    #norm
    nor = np.linalg.norm(x - z * aHat) ** 2
    return (aHat - z2 * np.sqrt(a_ii / (n - 1.)) * nor,
            aHat - z1 * np.sqrt(a_ii / (n - 1.)) * nor)
```

Подсчет интервалов для всех $n$: 

```python
y0 = []
y1 = []
for n in range(2, N + 1):
    (a1, a2) = leastSquaresA(samples[:n], n)
    y0.append(np.array(a1)[0][0])
    y1.append(np.array(a2)[0][0])
```

Строим график:

```python
nrange = range(2, N + 1)
plt.fill_between(nrange, y0, y1)
plt.plot(nrange, (lambda y: 0 * y)(np.array(nrange)), color='red')
plt.xlim(0, N)
plt.ylim(-10, 10)
plt.show()
```

#### Теперь представим, что мы знаем $\sigma^2 = 1$

Тогда зная: $$\frac{\hat{a} - a}{\sigma \sqrt{a_{ii}}} \sim \mathcal{N}(0, 1)$$
, где $a_{ii}$ - элемент матрицы $(Z^TZ)^{-1}$

Получим:

Получим:

$$P\left(\hat{a} - z_{\frac{1 + \alpha}{2}}\sigma\sqrt{a_{ii}} < a < \hat{a} - z_{\frac{1 - \alpha}{2}}\sigma\sqrt{a_{ii}}\right) = \alpha$$

Подсчет интервала:

```python
def leastSquaresAKnowSigma(samples, n):
    x = np.matrix(samples).T
    #квантили
    norm = sts.norm(loc=0, scale=1)
    z1 = norm.ppf(1. / 2. - alpha / 2.)
    z2 = norm.ppf(1. / 2. + alpha / 2.)
    #Z
    z = [1 for i in range(n)]
    z = np.matrix(z).T
    #a_ii
    a_ii = np.array(np.linalg.inv(z.T * z))[0][0]
    #aHat
    aHat = np.linalg.inv(z.T * z) * z.T * x
    return (aHat - z2 * np.sqrt(a_ii),
            aHat - z1 * np.sqrt(a_ii))
```

Подсчет интервалов для всех $n$: 

```python
y0 = []
y1 = []
for n in range(2, N + 1):
    (a1, a2) = leastSquaresAKnowSigma(samples[:n], n)
    y0.append(np.array(a1)[0][0])
    y1.append(np.array(a2)[0][0])
```

Строим график:

```python
nrange = range(2, N + 1)
plt.fill_between(nrange, y0, y1)
plt.plot(nrange, (lambda y: 0 * y)(np.array(nrange)), color='red')
plt.xlim(0, N)
plt.ylim(-2, 2)
plt.show()
```

Когда мы знаем сигму доверительный интервал получается гораздо лучше. (меньше)

#### Теперь представим, что мы знаем $a = 0$

$$P\left(\frac{\sum{(X_i - a)^2}}{\mathcal{X}^2_{\frac{1 + \alpha}{2}, n}} < \sigma^2 < \frac{\sum{(X_i - a)^2}}{\mathcal{X}^2_{\frac{1 - \alpha}{2}, n}}\right) = \alpha$$

Подсчет интервала:

```python
def leastSquaresSigmaKnowA(samples, n):
    #квантили
    norm = sts.chi2(df=n)
    z1 = norm.ppf(1. / 2. - alpha / 2.)
    z2 = norm.ppf(1. / 2. + alpha / 2.)
    return (np.array(samples ** 2).mean() * n / z2, np.array(samples ** 2).mean() * n / z1)
```

```python
y0 = []
y1 = []
for n in range(2, N + 1):
    (a1, a2) = leastSquaresSigmaKnowA(samples[:n], n)
    y0.append(a1)
    y1.append(a2)
```

```python
nrange = range(2, N + 1)
plt.fill_between(nrange, y0, y1)
plt.plot(nrange, (lambda y: 1 * y / y)(np.array(nrange)), color='red')
plt.xlim(0, N)
plt.ylim(0, 2)
plt.show()
```

Как мы видим результат снова улучшился. (доверительный интервал меньше)

#### Доверительная область для $(a, \sigma ^ 2)$

$P(0 < \sigma^2 < \frac{ns^2}{z_{1 - \sqrt{\alpha}}} \land \overline{X} - \sqrt{\frac{s^2z_{\sqrt{\alpha}}}{z_{1 - \sqrt{\alpha}}}} < a < \overline{X} + \sqrt{\frac{s^2z_{\sqrt{\alpha}}}{z_{1 - \sqrt{\alpha}}}}) = \alpha$
, где $z$ - квантили распределения $\mathcal{X}_{n - 1}^2$

Считаем область:

```python
def leastSquaresASigma(samples, n):
    samples = np.array(samples)
    #квантили
    norm = sts.chi2(df=(n - 1.))
    z1 = norm.ppf(np.sqrt(alpha))
    z2 = norm.ppf(1. - np.sqrt(alpha))
    #mean
    mean = samples.mean()
    #s2
    s2 = (samples ** 2).mean() - mean ** 2
    return (mean - (s2 * z1 / z2) ** 0.5,
            mean + (s2 * z1 / z2) ** 0.5,
            0, 
            n * s2 / z2
           )
```

```python
fig = plt.figure()
fig = plt.figure(figsize=(16, 10))
ax = Axes3D(fig)

verts = []
for n in range(1, N):
    (a1, a2, sigma1, sigma2) = leastSquaresASigma(samples[:n], n)
    xs = [a1, a1, a2, a2]
    ys = [sigma1, sigma2, sigma2, sigma1]
    zs = [n, n, n, n]
    verts.append(list(zip(xs, ys, zs)))
    
ax.add_collection3d(Poly3DCollection(verts, alpha=0.3))
ax.plot([0, 0], [1, 1], [1, N], color='red')
ax.set_xlabel("x")
ax.set_xlim3d(-2, 2)
ax.set_ylabel("y")
ax.set_zlim3d(1, N)
ax.set_zlabel("z")
ax.set_ylim3d(0, 5)

plt.show()
```

Как видим красная линия ($a = 0, \sigma ^ 2 = 1$) попадает внутрь доверительных областей.

А для всех графиков красная линия тоже попадала внутрь интервалов. Как и должно быть =)
