СДЕЛАЙТЕ ВЫВОД : НУМПАЙ РАБОТАЕТ БЫСТРО!!!
Примените модель ко второй части выборки и посчитайте по ней среднеквадратичную ошибку.

```python
def theta_est(xt, yt, lamt):
    return np.linalg.inv(xt.T * xt - lamt * np.eye(4)) * xt.T * yt
```

```python
lamrange = np.arange(140., 150.01, 0.5)
sigmas = []
step = 100.
for lam in lamrange:
    thetas = []
    for i in nrange:
        subx = x[nrange != i]
        suby = y[nrange != i]
        thetas.append(theta_est(subx, suby, lam))
    sigma = np.array(np.var(thetas, axis=0))
    sigmas.append(sigma)
    if lam >= step:
        print lam,
        step += 5
```

```python
sigmas = np.array(sigmas).T[0]
#print sigmas
```

```python
for i in range(4):
    plt.figure(figsize=(15,6))
    plt.plot(lamrange, sigmas[i], label="{}".format(i))
    plt.xlabel("lambda")
    plt.ylabel("sigma")
    plt.legend()
    plt.show()
```

```python
optlam = lamrange[np.argmax(sigmas[0])]
print optlam
```
