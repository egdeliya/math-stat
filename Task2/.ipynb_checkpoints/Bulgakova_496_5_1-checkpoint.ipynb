```python
import scipy.stats as sps
import numpy as np
import math
import matplotlib.pyplot as plt
%pylab inline
```

Построим график плотности для случайного вектора $\xi = (\xi_{1}, \xi_{2}) \sim N(a, \Sigma), a = 0,   \Sigma = \begin{pmatrix} 10 & 8 \\ 8 & 10 \end{pmatrix}.$

```python
grid = np.mgrid[-7:7:0.05, -7:7:0.05]
density = np.array([[sps.multivariate_normal.pdf((grid[0, i, j], grid[1, i, j]), mean=[0, 0], cov=[[10, 8], [8, 10]])
                       for i in range(grid[0].shape[0])]
                      for j in range(grid[0].shape[1])])

plt.figure(figsize=(15, 4))

subplot(1, 2, 1)
plt.pcolormesh(grid[0], grid[1], density, cmap='Blues')
plt.xlim((np.min(grid[0]) + 0.2, np.max(grid[0]) - 0.2))
plt.ylim((np.min(grid[1]) + 0.2, np.max(grid[1]) - 0.2))
plt.title('Gaussian density')

subplot(1, 2, 2)
CS = plt.contour(grid[0], grid[1], density, [0.001, 0.005, 0.01, 0.02, 0.03])
plt.clabel(CS, fontsize=14, inline=1, fmt='%1.3f')
plt.xlim((np.min(grid[0]), np.max(grid[0])))
plt.ylim((np.min(grid[1]), np.max(grid[1])))
    
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(16, 10))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(grid[0], grid[1], density, rstride=1, cstride=1, cmap='Blues_r',
                       linewidth=0, antialiased=False)
ax.set_zlim(0, 0.025)
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```

$f_{\xi_1|\xi_2}(x|y) =  0,  f_{\xi_2}(y) = 0.$

```python
plt.figure(figsize=(10, 8))

y = np.array([-3, 0, 1, 5])
cond_ex = np.zeros(4)

for i in range(4):
    cond_ex[i] = sps.norm.expect(loc=0, scale=math.sqrt(10), lb=y[i], conditional=True) 

plt.plot(y, cond_ex, color="blue", label=r'$E(\xi_1|\xi_2 = y)$')

for i in range(4):
    ex[i] = sps.norm.expect(loc=0, scale=math.sqrt(10)) 

plt.plot(y, ex, color="red", label=r'$x = E\xi_1$')

plt.ylim((-1, 7))

    
plt.legend()
plt.show()
```
