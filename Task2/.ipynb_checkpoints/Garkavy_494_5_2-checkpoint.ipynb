```python
import numpy
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
from mpl_toolkits.mplot3d.axes3d import Axes3D
from sklearn.datasets import load_iris
from scipy.stats import multivariate_normal
%matplotlib inline
```

```python
left = -1
right = 9 # границы на графиках
data = load_iris()
iris = [[], [], []]
for i in range(len(data.target)):
    iris[data.target[i]].append(data.data[i])
iris = numpy.array(iris)
```

```python
def build_density(coords, iris_type):
    d =  numpy.array([[a[coords[0]], a[coords[1]]] for a in iris[iris_type]])
    exp = (d[:,0].mean(), d[:,1].mean())
    cov = [[((d[:,i] * d[:,j]).mean() - d[:,i].mean() * d[:,j].mean()) for j in (0, 1)] for i in (0, 1)]
    
    x = numpy.linspace(left, right, 1000)
    y = numpy.linspace(left, right, 1000)
    x, y = numpy.meshgrid(x, y)
    z = mlab.bivariate_normal(x, y, sigmax=numpy.sqrt(cov[0][0]), sigmay=numpy.sqrt(cov[1][1]),
                              mux=exp[0], muy=exp[1], sigmaxy=cov[0][1])
    
    fig = plt.figure(figsize = (20, 10))
    ax = fig.gca()
    
    l = plt.title(r'coordinates - $' + str(coords) + r'$, type - $' + str(iris_type) + r'$')

    p1 = plt.contour(x, y, z)
    p2 = plt.plot(d[:,0], d[:,1], 'o')
    
    plt.show()
```

```python
for coords in [(0, 1), (1, 3), (2, 3)]:
    for iris_type in [0, 1, 2]:
        build_density(coords, iris_type)
```

Спроецируем точки выборки каждого типа на плоскости с парами координат (0, 1), (1, 3), (2, 3). Для каждого из 9 случаев (3 пары координат, 3 компоненты смеси) оценим плотность распределения.

Найдем условные мат. ожидания:

$E(X|T \neq 0) = (6.262,  2.872,  4.906,  1.676)$

$E(X|T \neq 1) = (5.797,  3.196,  3.508,  1.135)$

$E(X|T \neq 2) = (5.471,  3.094,  2.862,  0.785)$

```python
def build_cond_density(coords, k):
    pnts = numpy.array([[a[coords[0]], a[coords[1]]] for a in iris[k]])
    
    not_k = [0, 1, 2]
    not_k.remove(k)
    d =  [numpy.array([[a[coords[0]], a[coords[1]]] for a in iris[iris_type]]) for iris_type in not_k]
    exp = [(d[0][:,0].mean(), d[0][:,1].mean()), (d[1][:,0].mean(), d[1][:,1].mean())]
    cov = [[[((d[0][:,i] * d[0][:,j]).mean() - d[0][:,i].mean() * d[0][:,j].mean()) for j in (0, 1)] for i in (0, 1)],
           [[((d[1][:,i] * d[1][:,j]).mean() - d[1][:,i].mean() * d[1][:,j].mean()) for j in (0, 1)] for i in (0, 1)]]
           
    x = numpy.linspace(left, right, 1000)
    y = numpy.linspace(left, right, 1000)
    x, y = numpy.meshgrid(x, y)
    z = (mlab.bivariate_normal(x, y, sigmax=numpy.sqrt(cov[0][0][0]), sigmay=numpy.sqrt(cov[0][1][1]),
                               mux=exp[0][0], muy=exp[0][1], sigmaxy=cov[0][0][1]) \
        + mlab.bivariate_normal(x, y, sigmax=numpy.sqrt(cov[1][0][0]), sigmay=numpy.sqrt(cov[1][1][1]),
                                mux=exp[1][0], muy=exp[1][1], sigmaxy=cov[1][0][1])) \
        / 2
    
    fig = plt.figure(figsize = (20, 10))
    ax = fig.gca()
    
    l = plt.title(r'coordinates - $' + str(coords) + r'$, k - $' + str(k) + r'$')

    p1 = plt.contour(x, y, z)
    
    plt.show()
```

```python
for coords in [(0, 1), (1, 3), (2, 3)]:
    for k in [0, 1, 2]:
        build_cond_density(coords, k)
```

Построили 9 графиков $p(X|I(T \neq k))(x |1)$ для разных k и пар координат ($(0, 1), (1, 3), (2, 3)$), как просили в задании.

```python
d = iris
vec_exp = [numpy.array([d[i][:,j].mean() for j in (0, 1, 2, 3)]) for i in (0, 1, 2)]
matr_cov = [[[((d[k][:,i] * d[k][:,j]).mean() - d[k][:,i].mean() * d[k][:,j].mean()) for j in (0, 1, 2, 3)]
             for i in (0, 1, 2, 3)]
            for k in (0, 1, 2)]

def classify(point):    
    z = [multivariate_normal.pdf(point, mean=vec_exp[j], cov=matr_cov[j]) for j in (0, 1, 2)]
    return numpy.argmax(z)
```

```python
data_classification = [classify(x) for x in data.data]
```

```python
mistakes = 0
for i in range(len(data_classification)):
    if data_classification[i] != data.target[i]:
        mistakes += 1
print(mistakes / len(data_classification))
```

Проклассифицируем точки методом $k = argmax(p(X|I(T = k))(x |1))$. Количество ошибок - 2 %.

```python
def build_classification(coords, data_classification):
    N = 200
    def f(xs, ys):
        answer = []
        for x in xs:
            for y in ys:
                pnt = [0, 0, 0, 0]
                pnt[coords[0]] = x
                pnt[coords[1]] = y
                answer += [classify(pnt)]
        answer = numpy.array(answer)
        answer.shape = (N, N)
        return answer
    
    fig = plt.figure(figsize = (20, 10))
    ax = fig.gca()
    
    x0 = numpy.linspace(left,right,N)
    y0 = numpy.linspace(left,right,N)
    x, y = numpy.meshgrid(x0, y0)
    z = f(x0, y0)
    c = plt.contourf(x, y, z, 3, alpha=.75, cmap='jet')

    l = plt.title(r'coordinates - $' + str(coords) + '$')

    ax.set_xlim(left, right)
    ax.set_ylim(left, right)
    
    plt.show()
```

```python
for coords in [(0, 1), (1, 3), (2, 3)]:
    build_classification(coords, data_classification)
```

Итак, классифицируем 4-мерное пространство на примере трех плоскостей - плоскости пар координат (0, 1), (1, 3), (2, 3), причем все остальные координаты равны 0. Это малоинформативно, но в письме с вопросом мне ответили, что нужно классифицировать именно эти плоскости. Ну ладно.
